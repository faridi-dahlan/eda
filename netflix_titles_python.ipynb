```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/smartdata/netflix_titles_raw.csv')
```

```python
df
```

```python
df.info()
```

```python
# Konversi nilai string menjadi objek datetime
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y')
```

```python
df.info()
```

```python
df['duration'].unique()
```

```python
#merubah durasi agar seragam
df['duration_min'] = df['duration'].apply(lambda x: int(x.split()[0]) if 'min' in x else int(x.split()[0]) * 480)
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
for i in df.columns:
  try:
    df[i] = df[i].str.lower()
  except :
    pass
```

```python
df
```

```python
df.info()
```
